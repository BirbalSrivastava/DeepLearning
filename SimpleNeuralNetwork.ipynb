## A simple neural network to simulate a function 3x+2

```
## imports
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

```
## A simple keras sequential model with just one unit
model=tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

```
## Optimizer is sgd
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```
## Created a sample data
x=np.array([-9, 0, 1, 2, 3, 4], dtype=float)
y=np.array([-25, 2, 5, 8, 11, 14], dtype=float)
```

```
model.fit(x,y, epochs=1000)
```

```
## Check if this works. Answer should be close to 272
print(model.predict([90]))
```
